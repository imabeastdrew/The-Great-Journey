```python
from transformers import pipeline
```

```python
# create a pipeline object taking task 'sentiment analysis'
# this pipeline does 3 things for us:
# 1. preprocesses text (applies tokenizer)
# 2. feeds the text to model / applies model
# 3. postprocesses text (shows result how we would expect it)
# post processing for sentiment analysis : will show us the label
# positive or negative
classifier = pipeline("sentiment-analysis")

res = classifier("I've been waiting for a HuggingFace course my whole life")
```
